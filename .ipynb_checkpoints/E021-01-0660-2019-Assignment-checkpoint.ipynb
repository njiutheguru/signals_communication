```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

1. Consider the signal $(x(t) = r(t) − 2r(t − 1) + r(t − 2))$ where $r(t)$ is the unit ramp.<br>
(a) Sketch $x(t)$ <br>
(b) If $x(t)$ is the input to a differentiator whose response to $x(t)$ is $y(t) = \frac{dx}{dt}(x)$.sketch $y(t)$.

<br>
1. \( r(t) \): A ramp function starting at \( t = 0 \). <br>
2. \( r(t - 1) \): A ramp function starting at \( t = 1 \). <br>
3. \( r(t - 2) \): A ramp function starting at \( t = 2 \). <br>
Thus, the combined signal \( x(t) \) is the sum of these functions with the appropriate coefficients.

```python
# Define the unit ramp function r(t)
def ramp(t):
    return np.maximum(0, t)
# Define the signal x(t)
def x(t):
    return ramp(t) - 2 * ramp(t - 1) + ramp(t - 2)

# Define time array
t = np.linspace(-1, 3, 1000)
# Calculate x(t)
x_t = x(t)
# Plotting x(t)
plt.figure(figsize=(10, 6))
plt.plot(t, x_t, label='$x(t) = r(t) - 2r(t-1) + r(t-2)$')
plt.title('Signal $x(t)$')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define the delta function representation
def delta(t, t0, amplitude=1):
    return amplitude * np.where(np.isclose(t, t0, atol=1e-3), 1, 0)
# Calculate y(t)
y_t = delta(t, 0, 1) - 2 * delta(t, 1, 1) + delta(t, 2, 1)
# Plotting y(t)
plt.figure(figsize=(10, 6))
plt.stem(t, y_t, use_line_collection=True)
plt.title('Signal $y(t) = \\frac{dx(t)}{dt}$')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

2. x(t) = u(t) − 2u(t − 2) + u(t − 4) where u(t) is the unit step. <br>
(a) Sketch $x(t)$ <br>
(b) Sketch $x(2t)$ <br>
(c) Sketch $x(t/2)$ <br>

```python
def x(t):
    return (t >= 0).astype(float) - 2 * (t >= 2).astype(float) + (t >= 4).astype(float)

# Define the time axis for plotting
t = np.linspace(-1, 6, 800)  # Adjust the range to cover the signal properly

# Evaluate x(t) for the given time axis
xt = x(t)
```

```python
# Plotting x(t)
plt.figure(figsize=(10, 4))
plt.plot(t, xt, label='$x(t)$')
plt.title('Sketch of $x(t)$')
plt.xlabel('t')
plt.ylabel('$x(t)$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.xticks(np.arange(-1, 7, step=1))
plt.yticks(np.arange(-2, 3, step=1))
plt.ylim(-2.2, 2.2)
plt.legend()
plt.show()
```

```python
# Plotting x(2t)
plt.figure(figsize=(10, 4))
plt.plot(t, x(2*t), label='$x(2t)$')
plt.title('Sketch of $x(2t)$')
plt.xlabel('t')
plt.ylabel('$x(2t)$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.xticks(np.arange(-1, 7, step=1))
plt.yticks(np.arange(-2, 3, step=1))
plt.ylim(-2.2, 2.2)
plt.legend()
plt.show()
```

```python
# Plotting x(t/2)
plt.figure(figsize=(10, 4))
plt.plot(t, x(t/2), label='$x(t/2)$')
plt.title('Sketch of $x(t/2)$')
plt.xlabel('t')
plt.ylabel('$x(t/2)$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.xticks(np.arange(-1, 7, step=1))
plt.yticks(np.arange(-2, 3, step=1))
plt.ylim(-2.2, 2.2)
plt.legend()
plt.show()
```

3. The response y(t) of an integrator to an input x(t) is given by: <br>

<br>
\[ y(t) = \int_{-\infty}^{t} x(\tau) \, d\tau \]

(a) Verify that the integrator is time invariant. <br>
(b) Determine and sketch $y(t)$ when $x(t) = u(t) − 2u(t − 2) + u(t − 4)$ where $u(t)$ is the
unit step.


1. A system is time-invariant if a time shift in the input results in an identical time shift in the output. Mathematically, if:
<br>
\( x(t) \to y(t) \)

<br>
then for any time shift \( t_0 \), \( x(t - t_0) \to y(t - t_0) \).


2. **Integral System**: Given:
<br>
   \[
   y(t) = \int_{-\infty}^{t} x(\tau) \, d\tau
   \]
   


3. **Time-Shifted Input**: If the input is shifted by:  $( t_0 $), the new input is $( x(t - t_0) $).


4. **Output for Shifted Input**: Compute the output for the shifted input:
<br>
   \[
   y_{\text{shifted}}(t) = \int_{-\infty}^{t} x(\tau - t_0) \, d\tau
   \]



5. **Change of Variable**: Let 
<br>
\( \sigma = \tau - t_0 \), then \( d\sigma = d\tau \), and \( \tau = \sigma + t_0 \):
   \[
   y_{\text{shifted}}(t) = \int_{-\infty}^{t} x(\sigma) \, d\sigma
   \]

6. **Adjusting Limits**: Since $( \sigma $) is a dummy variable, we can rewrite the integral with $( \sigma $) replaced by $( \tau $) and adjust the limits accordingly:
<br>
   \[
   y_{\text{shifted}}(t) = \int_{-\infty}^{t-t_0} x(\tau) \, d\tau = y(t - t_0)
   \]

<br>
Thus, the output \( y_{\text{shifted}}(t) \) for the time-shifted input \( x(t - t_0) \) is \( y(t - t_0) \), confirming that the system is time-invariant.


```python
# Define the piecewise function for y(t)
def y(t):
    if t < 0:
        return 0
    elif 0 <= t < 2:
        return t
    elif 2 <= t < 4:
        return 4 - t
    else:
        return 0

# Generate time values
t = np.linspace(-1, 5, 400)
# Vectorize y(t) to apply it over arrays
y_vectorized = np.vectorize(y)

# Calculate y(t) for each t
y_values = y_vectorized(t)

# Plot the results
plt.figure(figsize=(8, 4))
plt.plot(t, y_values, label='$y(t)$', color='b')
plt.axhline(0, color='k', linestyle='--')
plt.axvline(0, color='k', linestyle='--')
plt.xlabel('t')
plt.ylabel('$y(t)$')
plt.title('$y(t) = \\int_{-\\infty}^{t} x(\\tau) \, d\\tau$ for $x(t) = u(t) - 2u(t - 2) + u(t - 4)$')
plt.grid(True)
plt.legend()
plt.show()
```

4. Determine the energy of the signal $(x(t) = u(t) − u(t − 4)$ where $u(t)$ is the unit step.

To determine the energy of the signal $( x(t) = u(t) - u(t - 4) $), where $( u(t) $) is the unit step function:

1. **Signal**:
<br>
   \[
   x(t) = u(t) - u(t - 4)
   \]

</br>
2. <strong>Simplifyint the Signal</strong>:
<br>
   - The unit step function \( u(t) \) is 1 for \( t \geq 0 \) and 0 otherwise.
   
</br>

<br>
   - Thus,
     \[
     x(t) = 
     \begin{cases}
     1 & \text{for } 0 \leq t < 4 \\
     0 & \text{otherwise}
     \end{cases}
     \]
</br>

3. **Signal Energy Formula**:
<br>
   \[
   E_x = \int_{-\infty}^{\infty} |x(t)|^2 \, dt
   \]
</br>
4. **Evaluate the Integral**:
<br>
   - The signal \( x(t) \) is nonzero only from 0 to 4, so:
     \[
     E_x = \int_{0}^{4} |1|^2 \, dt = \int_{0}^{4} 1 \, dt = 4
     \]
</br>
Thus, the energy of the signal \( x(t) = u(t) - u(t - 4) \) is \( 4 \).
