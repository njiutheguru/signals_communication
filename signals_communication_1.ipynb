```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of t values
t = np.arange(0, 2, 0.1)

# Calculate the corresponding y values using the cosine function
y = np.cos(2 * np.pi * t)

# Plot the cosine function
plt.plot(t, y, color='green')

# Add a horizontal line at y=0 to represent the x-axis
plt.axhline(0, color='black', linewidth=0.5)

# Customize the plot
plt.title('Graph of $x(t) = \cos(2\pi t)$')
plt.xlabel('$t$ (radians)')
plt.ylabel('$x(t)$')

# Set the aspect of the plot to be equal
plt.gca().set_aspect('equal', adjustable='box')

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sampling interval
Ts = 1/4

# Define the range of n values
n = np.arange(0, 20)

# Calculate the corresponding x[n] values using the sampled cosine function
x_n = np.cos((n * np.pi) / 2)

# Plot the discrete-time signal
plt.stem(n, x_n, use_line_collection=True)

# Customize the plot
plt.title(r'Discrete-Time Signal $x[n] = \cos\left(\frac{n\pi}{2}\right)$')
plt.xlabel(r'$n$')
plt.ylabel(r'$x[n]$')
plt.grid(True)

# Display the plot
plt.show()
```
